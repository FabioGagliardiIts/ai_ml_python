```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.impute as ski
```

```python

df_names = [
"symboling",
"normalized-losses",
"make",
"fuel-type",               
"aspiration",               
"num-of-doors",             
"body-style",              
"drive-wheels",            
"engine-location",         
"wheel-base",               
"length",                   
"width",                    
"height",                   
"curb-weight",              
"engine-type",              
"num-of-cylinders",         
"engine-size",              
"fuel-system",              
"bore",                     
"stroke",                  
"compression-ratio",        
"horsepower",               
"peak-rpm",                 
"city-mpg",                 
"highway-mpg",             
"price"]


df = pd.read_csv("https://raw.githubusercontent.com/FabioGagliardiIts/ai_ml_python/main/Esercizi/Lezione_15/imports-85.csv",
                sep=',',
                names =df_names).replace(to_replace="?",value=np.NaN)


```

```python
display(df)
```

```python
df["normalized-losses"] = ski.SimpleImputer(missing_values=np.NaN, strategy="mean").fit_transform(df["normalized-losses"].to_numpy().reshape(-1,1))
df["num-of-doors"] = ski.SimpleImputer(missing_values=np.NaN, strategy="most_frequent").fit_transform(df["num-of-doors"].to_numpy().reshape(-1,1))
```
