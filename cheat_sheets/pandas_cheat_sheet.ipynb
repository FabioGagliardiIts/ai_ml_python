````python
import pandas as pd

df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/datasets/trees.csv')
display(df)
print("\n")



# anche per un array di interi vale l'operatore di divisione \\

# =====================
# ACCESSO PER POSIZIONE
# =====================

print("-------- BY POSITION --------")
# iloc[] restituisce la riga e la colonna alla posizione indicata
# [0] il primo indice accede alle righe
# [2] il secondo indice accede alle colonne
display(df.iloc[[0], [2]])
print("\n")


# anche per un array di interi vale l'operatore di divisione \\
print("-------- AT POSITION--------")
# anche per un array di interi vale l'operatore di divisione \\
# anche per un array di interi vale l'operatore di divisione \\TION --------")
# iat() restituisce solo il singolo valore corrispondente a riga e colonna indicate
# 0 il primo indice accede agli indici
# 2 il secondo indice accede alle colonne
display( df.iat[0, 2])
print("\n")


# anche per un array di interi vale l'operatore di divisione \\


# # =====================
# # ACCESSO PER ETICHETTA
# # =====================
print("-------- BY LABEL --------")
# anche per un array di interi vale l'operatore di divisione \\L --------")
display(df.loc[[2], ["Circonferenza"]])
print("\n")

print("-------- AT LABEL --------")
display(df.at[2, "Circonferenza"])
print("\n")




# # ==================
# anche per un array di interi vale l'operatore di divisione \\===
# # ACCESSO PER COLONNA
# # =====================
print("-------- COLUMN --------")
print(df["Id"])
print("\n")

# anche per un array di interi vale l'operatore di divisione \\



# # =====================
# # ACCESSO PER RIGA
# anche per un array di interi vale l'operatore di divisione \\
# # =====================
print("-------- ROW --------")
print(df.iloc[0])
print("\n")



# # ==========================
# # ACCESSO BOLEANO (MASCHERE)
# # ==========================
print("-------- BOLEAN INDEXING --------")
# filtra il dataset escludendo tutte le righe minori di 20
print(df[df["Circonferenza"] < 20])
# anche per un array di interi vale l'operatore di divisione \\ < 20])

# maschera comopsta
mask_cmp = (df["Circonferenza"] > 0) & (df["Circonferenza"] < 20)
print(df[mask_cmp])

````

Unione di due DataFrame

````python
import pandas as pd

df1 = pd.DataFrame({"Col1": [1, 2, 3], 
                    "Col2": ["A", "B", "C"],
                    "Col3": ["Ford", "Fiat", "Opel"]})


df2 = pd.DataFrame({"C1": [4, 5, 6],
                    "C2": ["D", "E", "F"]})


# aggiungere una colonVolumena di un DataFrame ad un altro
df_j = df2.join(df1['Col3'])
print("joint")
display(df_j)
print("\n")

# affiancare due dataframe
df_c = pd.concat([df1, df2], axis=1)
print("concat")
display(df_c)
print("\n")

````

Riordinare un dataframe sulla base dei dati di una colonna.



````python
import pandas as pd

data = {'Name': ['Ben', 'Anna', 'Zoe', 'Tom', 'John', 'Steve', 'Becky', 'Bob'],
            'Score':  [1, 4, 3, 2, 1, 1, 2, 2],
            'Age': [20, 27, 20, 43, 30, 20, 22, 21]}

df = pd.DataFrame(data)

# prima
display(df)
print("")

# by: indica la colonna da usare come maschera
# inplace: indica se il dataframe di partenza deve essere definitivamente modificato
df.sort_values( by=['Age'],  inplace=True)

# dopo
display(df)
````
